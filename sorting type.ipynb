```
def selection_sort(L):
  for i in range(0,len(L)):
    min_index = L[i:].index(min(L[i:]))
    L[i], L[min_index+i] = L[min_index+i], L[i]
    print(L)

L = [33, 1, 41, 5, 2, 53, 1]
print(L)
selection_sort(L)
print(L)
```

```
def Bubble_sort(L):
  for i in range(len(L)):
    for j in range(len(L[:-i])):
      if L[j] > L[j+1]:
        L[j], L[j+1] = L[j+1], L[j]
    print(L)
  return L

L = [30, 12, 5, 81, 4, 4,]

print(L)
print(Bubble_sort(L))
```
